```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_csv('../data/monthwise/jan_AP.csv', index_col=0)
item_name   = pd.read_csv('../data/itemlist.csv', index_col=0)
```

```python
item_name   = item_name[(item_name['DEPID'] != 65) & (item_name['DEPID'] != 21)]
item_name   = item_name[['SKU','KEYWORD']]

item_name.count()
```

```python
df = df.merge(item_name, on='SKU')
len(df)
```

```python
basket = (df
          .groupby(['TXNID', 'KEYWORD'])['QUANTITY']
          .sum().unstack().reset_index().fillna(0)
          .set_index('TXNID'))
```

```python
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.0002, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules = rules[rules['confidence'] >= 0.5].sort_values('lift', ascending=False)
rules['antecedent support'] = rules['antecedent support']*len(df)
rules['consequent support'] = rules['consequent support']*len(df)
rules['support'] = rules['support']
```

```python
rules.head(100)
```

```python
support=[]
uniquedict = {}
for x in rules.iterrows():
    if x[1]['support'] not in support:
        support.append(x[1]['support'])
        uniquedict[x[1]['support']]=x[1]['antecedents'].union(x[1]['consequents'])
    else:
        uniquedict[x[1]['support']] = uniquedict[x[1]['support']].union(x[1]['antecedents']).union(x[1]['consequents'])
        
```

```python
uniquedict
```

```python
support=[]
item=[]
for x in uniquedict:
    for i in uniquedict[x]:
        support.append([x][0])
        item.append([i][0])
basketlist = pd.DataFrame({"Support":support, "Item":item})
basketlist
```
