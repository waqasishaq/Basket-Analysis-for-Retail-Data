```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("../data/monthwise/jan_ML.csv")
data.head()
```

```python
comb = data
kmeans = KMeans(n_clusters=9)
kmeans.fit(comb[['COST','MARGIN','QUANTITY']])
comb['LABEL'] = kmeans.predict(comb[['COST','MARGIN','QUANTITY']])
```

```python
comb.head(5)
```

```python
combLabel = comb.groupby(['LABEL']).agg({'COST': 'mean', 'MARGIN': 'mean', 'QUANTITY': 'mean'}).reset_index()
combLabel.head(30)
```

```python
label = ['1','2','3','4','5','6','7','8','9']
combLabel2 = combLabel
combLabel2=combLabel2.sort_values('COST')
combLabel2['COST_L']=label
combLabel2=combLabel2.sort_values('MARGIN')
combLabel2['MARGIN_L']=label
combLabel2=combLabel2.sort_values('QUANTITY')
combLabel2['QUANTITY_L']=label
combLabel2
```

```python
quant = data[['QUANTITY']]
kmeans = KMeans(n_clusters=5)
kmeans.fit(quant)
quant['Q_Label'] = kmeans.predict(quant)
display(quant.head(5))
label = ['L','LM','M','MH','H']
quantLabel = quant.groupby(['Q_Label']).agg({'QUANTITY': 'mean'}).reset_index().sort_values("QUANTITY")
quantLabel['Q_CAT']=label
quantLabel.head(5)
```

```python
cost = data[['COST']]
kmeans = KMeans(n_clusters=5)
kmeans.fit(cost)
cost['C_Label'] = kmeans.predict(cost)
display(cost.head(5))
label = ['L','LM','M','MH','H']
costLabel = cost.groupby(['C_Label']).agg({'COST': 'mean'}).reset_index().sort_values("COST")
costLabel['C_CAT'] = label
costLabel.head(5)
```

```python
mar = data[['SKU','MARGIN']]
kmeans = KMeans(n_clusters=5)
kmeans.fit(mar[['MARGIN']])
mar['M_Label'] = kmeans.predict(mar[['MARGIN']])
display(mar.head(5))
label = ['L','LM','M','MH','H']
marLabel = mar.groupby(['M_Label']).agg({'MARGIN': 'mean'}).reset_index().sort_values("MARGIN")
marLabel['M_CAT'] = label
marLabel.head(5)
```
