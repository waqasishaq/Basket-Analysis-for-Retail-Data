```python
import numpy as np
import pandas as pd
from datetime import datetime, date
```

```python
completeDF = pd.read_csv('data/ML_Complete.csv', index_col=0)
completeDF['TXNDATE'] =  pd.to_datetime(completeDF['TXNDATE'], format='%Y-%m-%d %H:%M:%S')
completeDF.head()
```

```python
SEN_DF = completeDF[completeDF['SKU']==70631].groupby([completeDF.TXNDATE.dt.year,completeDF.TXNDATE.dt.month,completeDF.TXNDATE.dt.day,completeDF.TXNDATE.dt.hour]).agg({'SELLPRICE':'mean', 'QUANTITY':'sum'})
```
